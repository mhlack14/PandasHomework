```python
import pandas as pd
import numpy as np
import csv
```

```python
#Pathing
data = "purchase_data.csv"
data_read = pd.read_csv(data)
```

```python

data_read.columns
```

```python
data_read.head(100)
```

```python
#Player Count

unique_values = len(data_read["SN"].unique())
total_players = pd.DataFrame({"Player_Count": [unique_values]})
total_players.head()
```

```python
#Purchasing Analysis Unique Items and Avgerage Price

unique_items = len(data_read["Item ID"].unique())
print(unique_items)
AVG_Price = (data_read["Price"].mean())
print(AVG_Price)
```

```python
#Purchasing Analysis Number of Purchases and Total Revenue brought in

Total_Num_Purchases=len(data_read["Purchase ID"].unique())
total_Revenue= (data_read["Price"].sum())

print(Total_Num_Purchases)
print(total_Revenue)
```

```python
#Purchasing analysis Data Frames

Data_Frame_PA = pd.DataFrame({"Total Unique Items": [unique_items], 
                          "Average Price":[AVG_Price], 
                          "Number of Purchases":[Total_Num_Purchases], 
                          "Revenue":[total_Revenue]})

Data_Frame_PA
```

```python
#Gender Demographics: show the returns

Gender = data_read["Gender"].unique()
Gender
```

```python
#Gender Demographics: Get the counts and then total percentages for Male, Other,Non-Disclosed, Female
#Get the list showing SN first for ease purposes 

data_read_group = data_read.groupby("SN").first()
data_read_group
```

```python
#Gender Demographics

Males = data_read_group["Gender"].value_counts()["Male"]
Females = data_read_group["Gender"].value_counts()["Female"]
Other = data_read_group["Gender"].value_counts()["Other / Non-Disclosed"]


total_gender_count = Males + Females + Other


print("total male count is", Males)
print("total female count is",Females)
print("total other count is",Other)
total_gender_count

```

```python
Percent_male = Males / total_gender_count * 100
percent_female = Females / total_gender_count * 100
percent_other = Other / total_gender_count * 100

Gender_list = gender_list = data_read_group["Gender"].unique()

Percentages = (Percent_male, percent_female, percent_other)
Total_counts = (Males, Females, Other)

gender_info = pd.DataFrame({"Gender": Gender_list,
                            "Total Count": Total_counts,
                            "Total Percentages": Percentages})
gender_info
```

```python
#Purchasing Analysis by Gender

Gender_dataframe = data_read.groupby(["Gender"])

total_purchases_count = Gender_dataframe["Price"].count()
Average_purchase_mean = Gender_dataframe["Price"].mean()
total_purchases_sum = Gender_dataframe["Price"].sum()
Normal_price_mean = Gender_dataframe["Price"].sum() / Gender_dataframe["Price"].count()

gender_purchasing = pd.DataFrame({"Purchase Count": total_purchases_count,
                                  "Average Purchase Price": Average_purchase_mean,
                                  "Total Purchase Value": total_purchases_sum,
                                  "Average total purchase per person" : Normal_price_mean
                                   })

gender_purchasing
```

```python
#Age Demographics

age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 100]
Age_ranges = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

data_read["Age Range"] = pd.cut(data_read["Age"], age_bins, labels= Age_ranges)


age_ranges_df = data_read.groupby(["Age Range"])
#age_ranges_df 

age_ranges_df
```

```python
#Age Demographics

player_count_by_age = age_ranges_df["SN"].count()
Percent_players_age = (age_ranges_df["SN"].count() / total_gender_count) * 100


age_demographics_df = pd.DataFrame({"Percentage of Players": Percent_players_age,
                                    "Total Count": player_count_by_age
                                    })
age_demographics_df

```

```python
age_purchases= data_read.groupby(["Age Range"])
#age_purchase_grouped

purchase_count = age_purchases["SN"].count()
#purchase_cnt

Average_purchase_mean = age_purchases["Price"].mean()
Total_purchase_value_mean = age_purchases["Price"].sum()

Normal_purchase_price_mean = age_purchases["Price"].sum() / age_purchases["Price"].count()

age_purchase_df = pd.DataFrame({"Purchases Count": purchase_count,
                                "Average Purchase Price": Average_purchase_mean,
                                "Total Purchase Value" : Total_purchase_value_mean,
                                "Normalized Totals" : Normal_purchase_price_mean})
age_purchase_df
```

```python
#Top 5 Spenders Analysis

data_read_SNGroup = data_read.groupby(["SN"])

#Get calc column values
SNgroupPurchaseCount = data_read_SNGroup["SN"].count()
SNgroupPurchaseCount = SNgroupPurchaseCount.sort_values(ascending=False)
#sn_purch_cnt.head(10)

data_read_SNGroup
SNgroupPurchaseCount
```

```python
average_purchase_price = data_read_SNGroup["Price"].mean()
purchase_value_total = data_read_SNGroup["Price"].sum()

#Dataframe
purchase_analysis_df = pd.DataFrame({"Purchase Count" : SNgroupPurchaseCount,
                                  "Average Purchase Price" : average_purchase_price,
                                  "Total Purchase Value" : purchase_value_total})

purchase_analysis_df = purchase_analysis_df.sort_values("Purchase Count", ascending=False)
purchase_analysis_df.head(10)
```

```python
#Popular Items

data_read.columns
data_read_popGroup = data_read.groupby(["Item ID","Item Name"])

Item_count = data_read_popGroup["Item Name"].count()
Item_price = data_read_popGroup["Price"].unique()
Item_price_sum = data_read_popGroup["Price"].sum()

Item_count
Item_price
Item_price_sum

```

```python
TopTenPopular_ItemsDF = pd.DataFrame({"Purchase Count" : Item_count,
                                "Item Price" : Item_price,
                                "Total Purchase Value" : Item_price_sum})


TopTenPopular_ItemsDF_Assorted =TopTenPopular_ItemsDF.sort_values("Purchase Count", ascending=False)  
TopTenPopular_ItemsDF_Assorted.head(10)
```

```python
#Most Profitable Items
MostProfitableItem = pd.DataFrame({"Purchase Count" : Item_count,
                                "Item Price" : Item_price,
                                "Total Purchase Value" : Item_price_sum})

#UNSORTED
MostProfitableItem
```

```python
#SORTED
MostProfitableItemSORTED = MostProfitableItem.sort_values("Total Purchase Value", ascending=False)

MostProfitableItemSORTED.head(10)
```
